## Bring biomass data into dataframe

```python
biomass = pd.read_csv('.\Analysis\Biomass_N_merged_WheatMaize.csv',index_col=0)
biomass.ix[:,'SimulationName'] = [TreatIndex.ix[biomass.ix[x,'Plot'],'SimName'] for x in biomass.index]
BiomassData = biomass.ix[biomass.Crop=='Wheat',['SimulationName','HarvestDate','Organ','BiomassWeight','NitrogenPerc','NitrogenWeight']]
BiomassData.ix[:,'HarvestDate'] = pd.to_datetime(BiomassData.ix[:,'HarvestDate']) 
BiomassData.replace(to_replace='Rachis',value='Spike',inplace=True)
BiomassData.columns = ['SimulationName','Clock.Today','Organ','BiomassWeight','NitrogenPerc','NitrogenWeight']
BiomassData.set_index(['SimulationName','Clock.Today','Organ'],inplace=True)

BiomassMeans = BiomassData.groupby(level=['SimulationName','Clock.Today','Organ']).mean()
BiomassMeans.drop('NitrogenPerc',axis=1,inplace=True)

#sum up total N amounts for all organs
TotalAboveGroundN = BiomassMeans.ix[:,'NitrogenWeight'].groupby(level=['SimulationName','Clock.Today']).sum()
#for biomass on final harvest the organs are also duplicated in the straw value so need to exclude then sum
AreOrgans = BiomassMeans.index.isin(['Grain', 'Spike', 'Stem','Leaf','Senesced'],level=2)
TotalAboveGroundWt = BiomassMeans.ix[AreOrgans,'BiomassWeight'].groupby(level=['SimulationName','Clock.Today']).sum()
TotalAboveGround = pd.concat([TotalAboveGroundWt,TotalAboveGroundN],axis=1)
TotalAboveGround.columns = ['Wheat.AboveGround.Wt','Wheat.AboveGround.N']
TotalAboveGround = TotalAboveGround.divide(10) #divide by ten to convert from kg/ha to g/m2
ObsDataFrame.update(TotalAboveGround)

Organs = ['Grain', 'Spike', 'Stem']

for Org in Organs:
    Data = BiomassMeans.ix[BiomassMeans.index.isin([Org],level=2),:]/10 ##Divide by ten to convert from kg/ha to g/m2
    Data.columns = ['Wheat.'+Org+'.Wt','Wheat.'+Org+'.N']
    Data.ix[:,'Wheat.'+Org+'.Nconc'] = Data.ix[:,'Wheat.'+Org+'.N']/Data.ix[:,'Wheat.'+Org+'.Wt']
    Data.index = Data.index.droplevel(2)
    ObsDataFrame.update(Data)

Data = BiomassMeans.ix[BiomassMeans.index.isin(['Senesced'],level=2),:]/10 ##Divide by ten to convert from kg/ha to g/m2 
Data.columns = ['Wheat.Leaf.Dead.Wt','Wheat.Leaf.Dead.N']
Data.ix[:,'Wheat.Leaf.Dead.Nconc'] = Data.ix[:,'Wheat.Leaf.Dead.N']/Data.ix[:,'Wheat.Leaf.Dead.Wt']
Data.index = Data.index.droplevel(2)
ObsDataFrame.update(Data)

Data = BiomassMeans.ix[BiomassMeans.index.isin(['Leaf'],level=2),:]/10 ##Divide by ten to convert from kg/ha to g/m2 
Data.columns = ['Wheat.Leaf.Live.Wt','Wheat.Leaf.Live.N']
Data.ix[:,'Wheat.Leaf.Live.Nconc'] = Data.ix[:,'Wheat.Leaf.Live.N']/Data.ix[:,'Wheat.Leaf.Live.Wt']
Data.index = Data.index.droplevel(2)
ObsDataFrame.update(Data)
```

## harvest data

```python
Harvest = pd.read_excel('K:\Rainshelter\Wheat 2015-16\Rainshelter Wheat 2015-2106 Phenology, Biomass, & Bird Damage.xlsx',
                           sheetname = 'Final Harvest Data',skiprows=[0,1,2,3])
HarvestData = Harvest.ix[:,['Harvest date',
                            'Plot',
                            '1000 grain weight (g)',
                            'Grain Yield @ 0% moisture (kg/ha)',
                            'Stem number (stem/m2)']]
HarvestData.ix[:,'Plot'] = HarvestData.ix[:,'Plot']-100
HarvestData.ix[:,'SimulationName'] = [TreatIndex.ix[HarvestData.ix[x,'Plot'],'SimName'] for x in HarvestData.index]
HarvestData.ix[:,'Harvest date'] = pd.to_datetime(HarvestData.ix[:,'Harvest date'])
HarvestData.drop('Plot',axis=1,inplace=True)
HarvestData.columns = ['Clock.Today',
                    'TGW',
                    'dryYield',
                    'Wheat.Structure.TotalStemPopn',
                    'SimulationName']
HarvestData.ix[:,'Wheat.Grain.GrainSize'] = HarvestData.ix[:,'TGW']/1000
HarvestData.ix[:,'Wheat.Grain.Number'] = HarvestData.ix[:,'dryYield']/10/HarvestData.ix[:,'Wheat.Grain.GrainSize']
HarvestData.drop(['dryYield','TGW'],axis=1,inplace=True)
HarvestData.set_index(['SimulationName','Clock.Today'],inplace=True)
HarvestMeans = HarvestData.groupby(level=['SimulationName','Clock.Today']).mean()
HarvestMeans.ix[:,'Wheat.Phenology.CurrentStageName'] = 'HarvestRipe'
ObsDataFrame.update(HarvestMeans)
```
